### 3.1.4 데이터프레임 인덱싱

```python
file_url = 'https://media.githubusercontent.com/media/musthave-ML10/data_source/main/sample_df.csv'
```

```python
sample_df = pd.read_csv(file_url)
```

```python
sample_df
```

index_col=n : n번째 열 기준으로 인덱싱

```python
sample_df = pd.read_csv(file_url, index_col=0)
```

```python
sample_df
```

한줄짜리 데이터는 시리즈 형태로 출력된다. 시리즈는 데이터프레임과는 다르게 위에 컬럼명이 표기되지 않는다.

```python
sample_df['var_1']
```

```python
sample_df[['var_1', 'var_2']]
```

컬럼을 인덱싱할 때 [ ]안에는 단 하나의 값만 넣을 수 있다. 따라서 리스트 []를 이용해서 [var_1, var_2]와 같이 만들면 하나의 값으로 인식되게 만들 수 있다.

```python
sample_df[['var_1', 'var_2']]
```

행 인덱스 이름 기준으로 인덱싱하기 : loc[] 이용

```python
sample_df.loc['a']
```

```python
sample_df.loc[['a','b','c']]
```

'a':'c' 자체가 하나의 값이므로 리스트를 사용할 필요 없음

```python
sample_df.loc['a':'c']
```

행 위치를 기준으로 인덱싱 : iloc[] 사용

```python
sample_df.iloc[[0,1,2]]
```

```python
sample_df.iloc[0:2]
```

```python
sample_df.iloc[0:3]
```

iloc[] 에서는 행 뿐만 아니라 컬럼까지 동시에 인덱싱할 수 있음

```python
sample_df.iloc[0:3, 2:4]
```

컬럼을 제거하려면 drop('컬럼명', axis=1) axis=1 지정해야 한다

```python
sample_df.drop('var_1', axis=1)
```

```python
sample_df.drop(['var_1','var_2'], axis=1)
```

```python
sample_df.drop(['a','b','c'])
```

### 3.1.9 데이터프레임 합치기

```python
left_url = 'https://media.githubusercontent.com/media/musthave-ML10/data_source/main/left.csv'
right_url = 'https://media.githubusercontent.com/media/musthave-ML10/data_source/main/right.csv'

left = pd.read_csv(left_url)
right = pd.read_csv(right_url)
```

```python
left
```

```python
right
```

merge() : 기본적으로 내부조인, 자동으로 공통 컬럼명 찾아 키값으로 활용

특정 변수를 지정해 키값으로 활용하고 싶으면 on 매개변수

```python
left.merge(right)
```

```python
left.merge(right, how = 'outer')
```

```python
left.merge(right, how = 'left')
```

join() : 기본적으로 왼쪽조인

인덱스를 기준으로 합치기 때문에, 컬럼명이 겹치면 합칠 수 없음. 특정 컬럼을 기준으로 join 하려면 컬럼을 인덱스로 설정해주어야 함 (set_index)

```python
left.join(right)
```

```python
left.drop('key', axis=1).join(right.drop('key', axis=1))
```

```python
left = left.set_index('key')
right = right.set_index('key')
```

```python
left
```

```python
left.join(right)
```

```python
left.join(right, how='inner')
```

```python
pd.concat([left,right])
```

```python
pd.concat([left,right], axis=1)
```
